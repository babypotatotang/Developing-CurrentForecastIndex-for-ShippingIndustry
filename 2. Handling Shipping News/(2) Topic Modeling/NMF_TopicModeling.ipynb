```python
import openpyxl
import pandas as pd
import re
from sklearn.feature_extraction.text import CountVectorizer,TfidfTransformer
from sklearn.decomposition import NMF
from sklearn.preprocessing import normalize
from gensim import corpora, models
```

```python
#날짜별 수집을 위한 Dates 리스트 준비(2000.01~2021.10)
read_file="C:\\Users\\User\\Desktop\\뉴스데이터 수집\\RawData\\Date_Filter\\Date.xlsx"

before=pd.read_excel(read_file,usecols=[0])
Before=before.values.tolist()
Before=sum(Before,[]) #날짜 데이터 준비
Before=list(map(str,Before))

after=pd.read_excel(read_file,usecols=[1])
After=after.values.tolist()
After=sum(After,[]) #날짜 데이터 준비
After=list(map(str,After))
```

```python
def get_topics(models,num):
    #the word ids obtained need to be reverse-mapped to the words so we can print the topic names.
    feat_names = vectorizer.get_feature_names()
    
    word_dict = {};
    for i in range(num):
        
        #for each topic, obtain the largest values, and add the words they map to into the dictionary.
        words_ids = model.components_[i].argsort()[:-10 - 1:-1]
        words = [feat_names[key] for key in words_ids]
        word_dict['Topic # ' + '{:02d}'.format(i+1)] = words;
    
    return pd.DataFrame(word_dict);
```

```python
#데이터셋 준비
document=[]

for i in range(17,262):
    read_file="C:\\Users\\User\\Desktop\\뉴스데이터 수집\\RawData\\Filtered\\NewsResult_Filtered\\NewsResult_"+Before[i]+"-"+After[i]+".xlsx"
    
    data=pd.read_excel(read_file,usecols=[2])
    tmp_document=data.values.tolist()
    tmp_document=sum(tmp_document,[])

    document=document+tmp_document
    
print("Document Ready")
```

```python
docs=pd.DataFrame(document)
docs.to_excel('total news.xlsx')
```

```python
text=document
vectorizer=CountVectorizer(analyzer='word')
x_counts=vectorizer.fit_transform(text)

transformer=TfidfTransformer(smooth_idf=False)
x_tfidf=transformer.fit_transform(x_counts)
```

```python
xtfidf_norm=normalize(x_tfidf,norm='l2',axis=1)

print("xtfidf_norm Ready")
#obtain a nmf model
model=NMF(n_components=80,init='nndsvd')
model.fit(xtfidf_norm)

print("Model Ready")
```

```python
print('tf-idf Dimensions '+str(x_tfidf.shape))
print('Feature Dimensions '+str(xtfidf_norm.shape))
print('Components Dimensions '+str(model.components_.shape))
```

```python
# Create a DataFrame: components_df
components_df = pd.DataFrame(model.components_, columns=vectorizer.get_feature_names())
```

```python
for topic in range(components_df.shape[0]):
    tmp = components_df.iloc[topic]
    
    print(f'For topic {topic+1} the words with the highest value are:')
    
    print(tmp.nlargest(25))
    print(type(tmp.nlargest(10)))
    print('\n')
    
```

```python
result=pd.DataFrame()
for topic in range(components_df.shape[0]):
    tmp = components_df.iloc[topic]
    tt=tmp.nlargest(20)
    tmp_df=pd.DataFrame(tt)
    result=result.append(tmp_df)
```

```python
result=result.transpose()
result.to_excel('tmp.xlsx')
```
